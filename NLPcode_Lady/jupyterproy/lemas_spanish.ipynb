```python
import nbimporter
from MySQL_connection import *
import spacy
```

```python
sql = "SELECT o_en FROM planes WHERE p = 'asignatura'"
lista_materias = []
sql_materias = ieee_select_query(sql)
for mat in sql_materias:
    lista_materias.append(mat['o'])
    
#print (lista_materias)
```

```python
nlp = spacy.load('es_core_news_md')
for asig in lista_materias:
    doc = nlp(asig)
    print ("---"+str(doc)+"---")
    for token in doc:
        print(token.text, token.lemma_ ,token.is_stop)
    print ("**************")
```
