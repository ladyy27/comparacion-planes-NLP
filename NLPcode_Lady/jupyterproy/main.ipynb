```python
#Capturar listas de stopwords
es_stopwords= stopwordsList("stopwords/spanish")
fr_stopwords= stopwordsList("stopwords/french")
en_stopwords= stopwordsList("stopwords/english")
punt_stopwords= stopwordsList("stopwords/puntuacion")
```

```python
#Variables de idioma
lang_es = "es"
lang_en = "en"
lang_fr = "fr"
```

```python
#RDF: Definir grafo
graph = create_graph()
```

```python
#Capturar el resultado de una consulta SQL
sql = "SELECT * FROM 2017_IEEE_Taxonomy_v1"
resultados = ieee_select_query(sql)
for res in resultados:
    print (res)
```

```python
#Capturar conceptos, procesamiento NLP y generar tripletas
idsentence = 1
idtoken = 1

for fila in resultados:
    concept = str(fila['Concept'])
    sentenceNode = add_node(sentence + "Sentence" + str (idsentence))
    add_graph(graph, sentenceNode, RDF.type, sentencetype)
    add_graph(graph, sentenceNode, sentenceId, Literal(idsentence, lang="es"))
    add_graph(graph, sentenceNode, dataSentence, Literal(concept, lang="es"))
    
    #Reemplazar tilde en textos en espaniol
    conceptSinTildes = replace_tildes(concept)
    text_object= TextBlob(conceptSinTildes)
    text_str = str(text_object)
    
    #Para capturar el error de Textblob que no permite traducir cadenas de menos de 2 caracteres
    if len(text_str) > 2:
        if text_object.detect_language() == lang_en:
            doc = en_parsing(text_str)
            generateTriplesByDetectedLang(doc,graph,sentenceNode,lang_en, idtoken, idsentence, en_stopwords, punt_stopwords)
        elif text_object.detect_language() == lang_es:
            doc = es_parsing(text_str)
            generateTriplesByDetectedLang(doc,graph,sentenceNode,lang_es, idtoken, idsentence, es_stopwords, punt_stopwords)
        elif text_object.detect_language() == lang_fr:
            doc = fr_parsing(text_str)
            generateTriplesByDetectedLang(doc,graph,sentenceNode,lang_fr, idtoken, idsentence, fr_stopwords, punt_stopwords)
        else:
            print ("Idioma no detectado: "+  text_str)
    else:
        print("- de 3 caracteres: " + text_str)
```
